```python
def convert_to_minutes(duration):
    """ 
    Converts a string representing duration (e.g., "1 hr., 30 min.") to minutes.

    Args:
        duration: The string representing the duration.

    Returns:
        The duration in minutes as an integer.
    """

    # Strip the blank spaces
    duration = duration.replace(" ", "")
    duration = duration.replace(".", "")

    if ("min" in duration) and ("hr" not in duration) and ("perep" not in duration):
        duration = duration.replace("min", "")
        minutes = duration
    elif ("min" in duration) and ("hr" in duration):
        duration = duration.replace("perep" , "")
        hrs = duration.split("hr")[0]
        min = duration.split("hr")[1].replace("min", "")
        minutes = str(int(hrs)*60 + int(min))
    elif ("perep" in duration):
        duration = duration.replace("minperep" , "")
        minutes = duration
    elif "sec" in duration:
        minutes = "-1"
    else:
        minutes = "-1"
    return minutes
```

```python
def Make_Plot_Bivariate(column1 , column2):
    fig, ax = plt.subplots(nrows=1, ncols=2 , figsize = (10 , 10))  # ax is now an array of Axes objects

    sns.scatterplot(x = column1, y= column2, ax=ax[0]) # Use ax[0]
    sns.lineplot(x=column1, y=column2, ax=ax[1])  # Use ax[1]

    # Set titles for each subplot
    ax[0].set_title('Scatter Plot')
    ax[1].set_title('Line Plot')
```

```python
Make_Plot_Bivariate(df["rank"] , df["Popularity"])
```

### Negative Correlation: 
Both plots clearly show a negative correlation between rank and popularity. As rank increases (gets worse), popularity tends to decrease. This is an expected and logical relationship - things that are ranked higher are generally more popular.
#### Non-linear Relationship: 
The relationship isn't strictly linear. The decrease in popularity is much steeper at the higher ranks (better ranks) and then seems to level off somewhat as the rank worsens.

#### Steep Decline Initially: 
The line plot visually emphasizes the steep decline in popularity as rank increases from 0 to around 2000-4000. This shows that the difference in popularity between the top-ranked items is substantial.
#### Flattening Curve: 
Beyond the initial steep decline, the curve flattens out, indicating that the difference in popularity between items ranked, say, 8000 and 12000, is less pronounced than the difference between items ranked 100 and 1000.
Volatility: The line plot also shows some volatility, especially at the lower ranks. This suggests fluctuations in popularity among items in those ranks.
